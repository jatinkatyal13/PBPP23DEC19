## WAP to copy a file to another location

```python
def copy(source, destination):
    with open(destination, "w") as dest, open(source, "r") as sour:
#             sour.write(desc.read())
        while True:
            temp = sour.read(256)
            if temp == '':
                break
            dest.write(temp)
```

```python
from functools import reduce
```

```python
arr = [1, 2, 3, 4, 5]
```

```python
students = [
    {
        "name":"jatin",
        "marks": 60
    },
    {
        "name":"himank",
        "marks": 90
    },
    {
        "name":"abhishek",
        "marks": 100
    },
    {
        "name":"himanshu",
        "marks": 80
    }
]
```
